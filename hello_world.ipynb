```
spam = {'color':'red', 'age':42}
print(spam)

for item in spam.items():
    print(item)

for key in spam.keys():
    print(key)

for value in spam.values():
    print(value)


for k, v in spam.items():
    print(k)
    print(v)


print('color' in spam.keys())

print('red' in spam.values())

print(spam.get('color'))
```
